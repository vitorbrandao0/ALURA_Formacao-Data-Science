# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor Mensal'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor Anual'] = dados['Valor Mensal']*12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'] + ' quartos ' + ' e ' + dados['Vagas'] + \
                     'vaga(s) de garagem.'
```

```python
dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quartos ' + ' e ' + dados['Vagas'].astype(str) + \
                     ' vaga(s) de garagem.'

dados.head()
```

```python
dados = dados.drop('Descricao', axis=1)
dados.head()
```

```python
dados['Possui Suítes?'] = dados['Suites'].apply(lambda x: "Sim" if x>0 else "Não")
dados.head()
```
